```python
import numpy as np
import pandas as pd
from copy import deepcopy
```

```python
train_df = pd.read_csv("train.csv")
```

```python
train_df.shape
```

```python
train_df.head()
```

```python
train_df["category"].value_counts()
```

```python
train_df = deepcopy(train_df.loc[train_df["category"] != "Report Unlawful Content"])
```

```python
train_df.shape
```

```python
labels = train_df['category'].unique().tolist()
labels = [s.strip() for s in labels ]
labels
```

```python
NUM_LABELS = len(labels)

id2label = {id:label for id,label in enumerate(labels)}

label2id = {label:id for id,label in enumerate(labels)}
```

```python
label2id
```

```python
id2label
```

```python
train_df["labels"] = train_df.category.map(lambda x: label2id[x.strip()])
```
